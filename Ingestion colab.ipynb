```python
import os
import h5py
import numpy as np
import pandas as pd

def load_cleaned_light_curve(file_path):
    """
    Loads cleaned time-series starlight data from an HDF5 file.
    """
    if not os.path.exists(file_path):
        raise FileNotFoundError(f"Target data file not found at: {file_path}")

    with h5py.File(file_path, 'r') as f:
        time = np.array(f['time'])
        flux = np.array(f['flux'])
        star_id = f.attrs.get('star_id', 'Unknown_Target')

    valid_mask = ~np.isnan(time) & ~np.isnan(flux)
    time = time[valid_mask]
    flux = flux[valid_mask]

    df = pd.DataFrame({
        'time': time,
        'flux': flux
    })

    print(f"[SUCCESS] Loaded target {star_id}: {len(df)} data points remaining after masking.")
    return df
```
