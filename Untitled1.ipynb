```python
import pandas as pd
import numpy as np
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix

# Set a random seed so we get the same "random" numbers every time
np.random.seed(42)

# Let's generate 1,000 fake transit events
num_samples = 1000

# Generating fake features based on what Role 2 is supposed to deliver
data = {
    # Transit Depth: How much the starlight dipped (usually tiny, 0.1% to 5%)
    'transit_depth': np.random.uniform(0.001, 0.05, num_samples),

    # Transit Duration: How long the dip lasted in hours
    'transit_duration': np.random.uniform(1.0, 8.0, num_samples),

    # Ingress/Egress Time: How long it took the planet to cross the edge of the star
    'ingress_egress_time': np.random.uniform(0.1, 0.5, num_samples),

    # Impact Parameter: How close to the center of the star the planet crossed (0 to 1)
    'impact_parameter': np.random.uniform(0.0, 1.0, num_samples),

    # Signal-to-Noise Ratio (SNR): How clear the signal is
    'snr': np.random.uniform(5.0, 50.0, num_samples),

    # Background Offset: Did the center of light shift? (Often indicates a false positive)
    'centroid_offset': np.random.uniform(0.0, 2.0, num_samples),
}

# Create the DataFrame
df = pd.DataFrame(data)

# Now, let's create the Target column: 'is_exoplanet' (1 = Planet, 0 = False Positive/Noise)
# We'll make it realistic: mostly 0s (space is empty!), with about 15% actual planets.
# We will also add a slight pattern so the ML model actually has something to learn:
# Planets usually have lower centroid offsets and distinct depth/duration ratios.

conditions = (
    (df['centroid_offset'] < 0.5) &
    (df['snr'] > 10.0) &
    (np.random.rand(num_samples) > 0.3) # Adding some randomness so it's not perfectly predictable
)

df['is_exoplanet'] = np.where(conditions, 1, 0)

df.to_csv('lightcurve_features.csv', index=False)

print(f"Success! Generated a dummy dataset with {len(df)} rows.")
print(f"Number of Exoplanets (1s): {df['is_exoplanet'].sum()}")
print(f"Number of False Positives (0s): {len(df) - df['is_exoplanet'].sum()}")
print("File saved as 'lightcurve_features.csv'. You can now run the XGBoost script!")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
import numpy as np

# Set device to GPU if available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# 1. Define the 1D-CNN Architecture
class Exoplanet1DCNN(nn.Module):
    def __init__(self, sequence_length):
        super(Exoplanet1DCNN, self).__init__()

        # Conv1D Layer 1: Slides a window of size 5 across the sequence
        # Why: It extracts low-level shape features (local slopes, micro-dips)
        self.conv1 = nn.Conv1d(in_channels=1, out_channels=16, kernel_size=5, stride=1, padding=2)
        self.relu1 = nn.ReLU()
        self.pool1 = nn.MaxPool1d(kernel_size=2) # Shrinks data size by half to save memory

        # Conv1D Layer 2: Looks at combinations of features found in layer 1
        # Why: It stitches small shapes together to see the wider "U-shape"
        self.conv2 = nn.Conv1d(in_channels=16, out_channels=32, kernel_size=5, stride=1, padding=2)
        self.relu2 = nn.ReLU()
        self.pool2 = nn.MaxPool1d(kernel_size=2)

        # Calculate the size of data entering the fully connected layers
        self.flatten_size = 32 * (sequence_length // 4)

        # Fully Connected Layers (The Classifier)
        # Why: Takes the structural shapes found by the convolutions and makes the final guess
        self.fc1 = nn.Linear(self.flatten_size, 64)
        self.relu3 = nn.ReLU()
        self.fc2 = nn.Linear(64, 1) # Outputs a single raw score (Logit)
        self.sigmoid = nn.Sigmoid() # Squashes score between 0 and 1 (Probability)

    def forward(self, x):
        x = self.pool1(self.relu1(self.conv1(x)))
        x = self.pool2(self.relu2(self.conv2(x)))
        x = x.view(x.size(0), -1) # Flatten for dense layers
        x = self.relu3(self.fc1(x))
        x = self.sigmoid(self.fc2(x))
        return x
```

```python
# 1. Generate Synthetic Sequential Data (1000 samples, sequence length 200)
num_samples = 1000
seq_length = 200

# Creating random noise timelines
X_fake = np.random.normal(0, 0.2, (num_samples, 1, seq_length)).astype(np.float32)
y_fake = np.random.randint(0, 2, (num_samples, 1)).astype(np.float32)

# Injecting fake "transits" (dips) into the sequences where y == 1
for i in range(num_samples):
    if y_fake[i] == 1:
        # Lower the brightness in the middle of the timeline to create a transit dip
        X_fake[i, 0, 80:120] -= 0.6

# 2. Convert to PyTorch Tensors and create Data Loaders
X_tensor = torch.tensor(X_fake)
y_tensor = torch.tensor(y_fake)

dataset = TensorDataset(X_tensor, y_tensor)
train_loader = DataLoader(dataset, batch_size=32, shuffle=True)

# 3. Initialize Model, Loss Function, and Optimizer
model = Exoplanet1DCNN(sequence_length=seq_length).to(device)
criterion = nn.BCELoss() # Binary Cross Entropy Loss (standard for 1/0 classification)
optimizer = optim.Adam(model.parameters(), lr=0.001) # Smart gradient descent

# 4. The Training Loop
# Why: This tells the network to look at data, guess, calculate error, and adjust internal weights.
epochs = 5
print("Starting 1D-CNN Training Loop...")

for epoch in range(epochs):
    model.train()
    running_loss = 0.0

    for batch_X, batch_y in train_loader:
        # Move batch data onto the Colab cloud GPU
        batch_X, batch_y = batch_X.to(device), batch_y.to(device)

        # Zero out old gradients from the previous turn
        optimizer.zero_grad()

        # Forward pass: model makes predictions
        outputs = model(batch_X)

        # Calculate how wrong the predictions were
        loss = criterion(outputs, batch_y)

        # Backward pass: compute how to tweak weights to fix errors
        loss.backward()

        # Step: Actually update the internal parameters
        optimizer.step()

        running_loss += loss.item() * batch_X.size(0)

    epoch_loss = running_loss / num_samples
    print(f"Epoch {epoch+1}/{epochs} - Loss: {epoch_loss:.4f}")

print("Phase 2 training script verified and operational!")
```

```python
import torch
import torch.nn as nn

class Astroformer(nn.Module):
    def __init__(self, sequence_length, feature_dim=1, num_heads=4, hidden_dim=64, num_layers=2):
        super(Astroformer, self).__init__()

        # 1. Input Projection: Scales our 1D brightness data up to a larger dimension
        self.input_projection = nn.Linear(feature_dim, hidden_dim)

        # 2. Positional Encoding: Transformers look at everything at once,
        # so we have to explicitly tell it the "time order" of the data.
        self.positional_encoding = nn.Parameter(torch.randn(1, sequence_length, hidden_dim))

        # 3. The Transformer Encoder Block (The actual brain)
        encoder_layer = nn.TransformerEncoderLayer(
            d_model=hidden_dim,
            nhead=num_heads,
            dim_feedforward=hidden_dim * 4,
            batch_first=True,
            dropout=0.1
        )
        self.transformer_encoder = nn.TransformerEncoder(encoder_layer, num_layers=num_layers)

        # 4. The Classifier Head
        self.fc_out = nn.Linear(hidden_dim, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        # x shape expects: [Batch Size, Sequence Length, Features (1)]
        # Notice this is slightly different from the CNN which expected [Batch, Channels, Length]

        # Project input to higher dimension and add positional info
        x = self.input_projection(x) + self.positional_encoding

        # Pass through the attention mechanism
        x = self.transformer_encoder(x)

        # We only care about the final prediction, so we pool the sequence (take the mean)
        x = x.mean(dim=1)

        # Final classification
        x = self.sigmoid(self.fc_out(x))
        return x

# Let's test if it fits on the GPU without crashing
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Note: Sequence length is 200, feature dim is 1 (brightness)
test_model = Astroformer(sequence_length=200).to(device)

# Create a fake batch of data formatted for the Transformer [Batch=32, Seq=200, Features=1]
dummy_input = torch.randn(32, 200, 1).to(device)
output = test_model(dummy_input)

print("Transformer architecture initialized successfully!")
print(f"Output shape (should be [32, 1]): {output.shape}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc, precision_recall_curve, classification_report

# 1. Simulate the probabilities from your trained models
# In reality, you will get these using: xgb_probs = xgb_model.predict_proba(X_test)[:, 1]
num_samples = 1000
y_true = np.random.randint(0, 2, num_samples) # The actual answers (1 = Planet, 0 = Noise)

# Let's pretend XGBoost is about 80% accurate, and the 1D-CNN is 85% accurate
xgb_probs = np.clip(y_true - np.random.normal(0, 0.4, num_samples) + 0.2, 0, 1)
cnn_probs = np.clip(y_true - np.random.normal(0, 0.3, num_samples), 0, 1)

# 2. The Fusion Layer (Weighted Soft Voting)
# We give the 1D-CNN slightly more voting power because it analyzes raw signal shapes
weight_xgb = 0.4
weight_cnn = 0.6

print("Fusing model predictions...")
fusion_probs = (xgb_probs * weight_xgb) + (cnn_probs * weight_cnn)

# 3. Final Decision Threshold
# If the combined probability is > 0.5, we declare it an Exoplanet
final_predictions = (fusion_probs > 0.5).astype(int)

print("\n--- Fusion Model Evaluation ---")
print(classification_report(y_true, final_predictions))

# 4. Generate the ROC Curve (For your final report!)
# ROC shows how well the model separates Planets from False Positives at various thresholds
fpr, tpr, thresholds = roc_curve(y_true, fusion_probs)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, lw=2, label=f'Fusion Ensemble (AUC = {roc_auc:.3f})')
plt.plot([0, 1], [0, 1], color='gray', lw=2, linestyle='--', label='Random Guessing')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate (Mistaking noise for a planet)')
plt.ylabel('True Positive Rate (Successfully finding a planet)')
plt.title('Receiver Operating Characteristic (ROC) - Exoplanet Detection')
plt.legend(loc="lower right")
plt.grid(alpha=0.3)
plt.show()
```
